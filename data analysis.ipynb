```python
# how to Define function in python
def sine(x):
    y = np.sin(x)
    return y
def cosine(x):
    y = np.cos(x)
    return y
c = np.linspace(1, 360, 100)
c = np.deg2rad(c)

y = sine(c)
y1 = cosine(c)
print(y)
```

```python
# Plotting the generated data from the previous cell

plt.style.use("ggplot")
plt.figure(figsize=(7,5))
plt.plot(c,y,color ="yellow",label="y=sin(x)")
plt.plot(c,y1,color = "purple",label="y=cos(x)")
plt.xlabel("x[degree]", size=10)
plt.ylabel("y",size=10)
plt.title("Sine Vs Cosine function",size = 18 , pad = 20)
plt.xlim(0,6)
plt.legend()
plt.grid
plt.show
```

```python
# Plotting the generated data from the previous cell

plt.style.use("ggplot")
plt.figure(figsize=(7,5))
plt.subplot(1,2,1)
plt.plot(c,y,color ="yellow",label="y=sin(x)")
#plt.plot(c,y1,color = "purple",label="y=cos(x)")
plt.xlabel("x[degree]", size=10)
plt.ylabel("y",size=10)
plt.title("sine function",size = 18 , pad = 20)
plt.xlim(0,6)
plt.legend()
plt.subplot(1,2,2)
plt.plot(c,y1,color = "purple")
plt.title("cosine function",size = 18,pad = 20)
plt.grid
plt.show
```
